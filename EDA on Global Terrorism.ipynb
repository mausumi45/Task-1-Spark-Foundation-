# Data Cleaning

```python
df = df[['iyear','imonth','iday','country_txt','region_txt','provstate','city','latitude','longitude','success',
         'attacktype1','attacktype1_txt','targtype1','targtype1_txt','weaptype1','weaptype1_txt','gname','nkill',
        'nwound','addnotes']]
df.head()
```

### Renaming the column

```python
df.rename(columns ={'iyear':'Year','imonth':'month','iday':'day','country_txt':'country','region_txt':'region',
                    'provstate':'state','city':'city','latitude':'latitude','longitude':'longitude','success':
                    'success','attacktype1':'attack','attacktype1_txt':'attack_category','targtype1':'target',
                    'targtype1_txt':'target_type','weaptype1':'weapon','weaptype1_txt':'weapon_type','gname':
                    'group_name','nkill':'killed','nwound':'wounded','addnotes' :'Add notes'},inplace = True)
```

```python
df.head()
```

```python
df.info()
```

### missing values

```python
df.isnull().sum()
```

### replacing null values of  killed and wounded columns by 0

```python
df['killed'] = df['killed'].fillna(0)
df['wounded'] = df['wounded'].fillna(0)
```

```python
df['casualty'] = df['killed']+df['wounded']
```

# Visualizing the data

## 1.Yearwise Analysis

## Number of attacks each year

```python
year = df['Year'].unique()
Number_Attacks = df['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (18,10))
sns.barplot(x = year,
           y = Number_Attacks,
           palette = 'terrain_r')
plt.xticks(rotation = 45)
plt.xlabel('Attack Year')
plt.ylabel('Number of Attacks each year')
plt.title('Attack_of_Years')
plt.show()
```

#### Highest number of attack happen in 2014(i.e. 16903) and the least number of attack in 1970(i.e. 475)

## Number of casualty each year

```python
yearwise_casuality = df[['Year','casualty']].groupby('Year').sum()
yearwise_casuality.plot(kind ='bar',color='#fb1',figsize = (18,10))
plt.xticks(fontsize = 12)
plt.xlabel('Year', fontsize =12)
plt.ylabel('Number of casualty each year',fontsize=13)
plt.title('number_of_casualty_each_Years')
plt.show()
```

## Number of people Killed each year

```python
yearwise_Killed = df[['Year','killed']].groupby('Year').sum()
yearwise_Killed.plot(kind ='bar',color='tab:blue',figsize = (18,10))
plt.xticks(rotation = 45)
plt.xlabel('Number of people died', fontsize =13)
plt.ylabel('Year',fontsize=13)
plt.title('number_of_people_died_each_Years')
plt.show()
```

## Number of Injured people each year

```python
yearwise_injured = df[['Year','wounded']].groupby('Year').sum()
yearwise_injured.plot(kind ='bar',color='#00FF00',figsize = (18,10))
plt.xticks(fontsize = 12)
plt.xlabel('Year', fontsize =12)
plt.ylabel('Number of injured people',fontsize=13)
plt.title('number_of_injured_people_each_Years')
plt.show()
```
